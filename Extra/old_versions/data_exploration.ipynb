```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/smart_manufacturing_data.csv')
# Plot distributions for sensor columns
for col in ['temperature', 'vibration', 'humidity', 'pressure', 'energy_consumption']:
    plt.figure()
    df[col].hist(bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
