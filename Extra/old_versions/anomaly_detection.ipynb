```
import pandas as pd
from sklearn.ensemble import IsolationForest
df = pd.read_csv('../data/smart_manufacturing_data.csv')
from feature_engineering import compute_features
features = compute_features(df)
X = features.drop(columns=['machine_id','anomaly_flag','maintenance_required','downtime_risk','predicted_remaining_life','failure_type'])
clf = IsolationForest(contamination=0.1, random_state=42)
preds = clf.fit_predict(X)
# Anomalies indicated by -1
print('Number of anomalies detected:', (preds==-1).sum())
```
