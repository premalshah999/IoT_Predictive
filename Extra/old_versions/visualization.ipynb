```
import pandas as pd
import matplotlib.pyplot as plt
# Load cross validation results
df_anom = pd.read_csv('../results/results_anomaly.csv')
df_maint = pd.read_csv('../results/results_maintenance.csv')
df_dt = pd.read_csv('../results/results_downtime.csv')
df_fail = pd.read_csv('../results/results_failure_type.csv')
df_rul = pd.read_csv('../results/results_remaining_life.csv')
# Plot F1 scores for each target
for name, df in [('Anomaly', df_anom),('Maintenance', df_maint),('Downtime', df_dt),('FailureType', df_fail)]:
    plt.figure()
    plt.bar(df['model'], df['f1_mean'])
    plt.title(f'F1 Scores for {name}')
    plt.xticks(rotation=45)
    plt.ylabel('F1 Score')
    plt.tight_layout()
    plt.show()
```
